# 複数の直線グラフ

```python
import matplotlib.pyplot as plt

def intro():

	plt.plot([15, 30])    # 青
	plt.plot([0.5, 1.5], [35, 25])    # オレンジ
	plt.plot([0, 0.5], [15, 35])    # 緑
	plt.show()

if __name__=='__main__':
	intro()
```

# タイトル、軸ラベル指定

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family="MS Gothic")

def intro():

	plt.plot([15, 30])
	plt.plot([35, 25])
	plt.title('2つの直線')
	plt.xlabel('x座標')
	plt.ylabel('y座標')
	plt.show()

if __name__=='__main__':
	intro()
```
