# 文字列の描画1_axes-text

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	fig = plt.figure(figsize=(10, 6))

	# 水平方向
	ax1 = fig.add_subplot(1,3,1)
	ax1.scatter([2, 2, 2, 2],[2, 4, 6, 8], color='r')
	ax1.text(2, 2, 'right',  ha='right',  fontsize=12)
	ax1.text(2, 4, 'center', ha='center', fontsize=12)
	ax1.text(2, 6, 'left',   ha='left',   fontsize=12)
	ax1.text(2, 8, 'text',                fontsize=12)
	ax1.grid()
	ax1.set(
	    title='水平方向', 
	    xlim=[0, 4], 
	    ylim=[1, 10]
	)

	# 垂直方向
	ax2 = fig.add_subplot(1,3,2, sharey=ax1)
	ax2.scatter([2, 2, 2],[4, 6, 8], color='r')
	ax2.text(2, 4, 'bottom', va='bottom', fontsize=12)
	ax2.text(2, 6, 'center', va='center', fontsize=12)
	ax2.text(2, 8, 'top',    va='top',    fontsize=12)
	ax2.grid()
	ax2.set(
	    title='垂直方向', 
	    xlim=[0, 4], 
	    ylim=[1, 10]
	)

	# 回転
	ax3 = fig.add_subplot(1,3,3, sharey=ax1)
	ax3.scatter([1, 2, 3],[2, 5, 8], color='r')
	ax3.text(1, 2, '90°bottom', rotation='vertical', va='bottom', fontsize=12)
	ax3.text(2, 5, '90°center', rotation='vertical', va='center', fontsize=12)
	ax3.text(3, 8, '90°top',    rotation='vertical', va='top',    fontsize=12)
	ax3.grid()
	ax3.set(
	    title='回転',
	    xlim=[0, 4],
	    ylim=[1, 10]
	)

	plt.show()

if __name__=='__main__':
	basic()
```

# 文字列の描画2_axes-text

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	fig = plt.figure(figsize=(8,6))
	ax = fig.add_subplot()
	
	bstyles = ['circle', 'darrow', 'larrow', 'rarrow', 'round',
	           'round4', 'roundtooth', 'sawtooth', 'square']

	for i, bstyle in enumerate(bstyles):
		ax.text(
			(i%3)*3+2, int(i/3)*3+1, bstyle,
			color='w', fontsize=20,
			bbox={
				"boxstyle": bstyle,
				"fc": 'C{}'.format(i)
				}
		)

	ax.set(
		xlim=(0, 10),
		ylim=(0, 8)
	)
	fig.suptitle('Box Styles')

	plt.show()

if __name__=='__main__':
	basic()
```

# 凡例のカスタマイズ_axes-legend

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	x = np.linspace(-2, 5, 71)  # 区間[-2, 5]を70等分する71点

	fig = plt.figure(figsize=(15, 10))

	# (1)基本
	ax = fig.add_subplot(3,4,1)
	ax.plot(x, np.sin(x), label='sin(x)', color='red')
	ax.legend()

	# (2)ラベルを別途与える
	ax = fig.add_subplot(3,4,2)
	line, = ax.plot(x, np.sin(x), color='red')
	line.set(label='sin(x)')
	ax.set(title='ラベルを別途外からセットする')
	ax.legend()

	# (3)ラベルを別途与える(リストで与える）
	ax = fig.add_subplot(3,4,3)
	ax.plot(x, np.sin(x), color='red')
	ax.plot(x, np.log(x), color='orange')
	ax.legend(['sin(x)', 'log(x)'])
	ax.set(title='ラベルを外からリストで与える')

	# (4)ラベルを更新する（タプルで与える）。凡例の文字の大きさ）
	ax = fig.add_subplot(3,4,4)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	line3, = ax.plot(x, np.sqrt(x), label='sqrt(x)', color='blue')
	ax.legend((line3, line2, line1), ('SQRT(X)', 'LOG(X)', 'SIN(X)'), 
	            title='凡例', title_fontsize=10, fontsize='x-small', loc=9)
	ax.set(title='ラベルを外からタプルで与える＋凡例文字カスタマイズ')

	# (5)選択したラベルのみ表示する
	ax = fig.add_subplot(3,4,5)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	line3, = ax.plot(x, np.sqrt(x), label='sqrt(x)', color='blue')
	ax.legend(handles=[line2, line1])
	ax.set(title='選択したラベルのみ表示する')

	# (6)選択したラベルを更新して表示する。（リストで与える）
	ax = fig.add_subplot(3,4,6)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	line3, = ax.plot(x, np.sqrt(x), label='sqrt(x)', color='blue')
	ax.legend([line1, line2], ['SIN（X)', 'SQRT(X)'])
	ax.set(title='選択したラベルを更新して表示')

	# (7)凡例を欄外上部に・・・
	ax = fig.add_subplot(3,4,7)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	ax.legend(
	    bbox_to_anchor=(0, 1.02, 1, .102),
	    ncol=2, mode="expand", borderaxespad=0.5)
	ax.set(title='上部に表示')

	# (8)凡例を欄外右に…
	ax = fig.add_subplot(3,4,8)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	ax.legend(bbox_to_anchor=(1.02, 1.0), borderaxespad=0)
	ax.set(title='右に表示')

	# (9)凡例を分けて表示
	ax = fig.add_subplot(3,4,9)
	line1, = ax.plot(x, np.sin(x), label='sin(x)', color='red')
	line2, = ax.plot(x, np.log(x), label='log(x)', color='orange')
	first_legend = ax.legend(handles=[line1], loc=1)
	# legend()を何回も実行すると最後のlegend()の結果しか残らない。
	# そのため、今書いたlegend()をadd_artist()で図中に加えておく。	ax.add_artist(first_legend)
	ax.legend(handles=[line2], loc=4)
	ax.set(title='分けて表示')

	plt.tight_layout()
	plt.show()

if __name__=='__main__':
	basic()
```

# 軸のカスタマイズ_axes-spines

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")


def basic():

	def plot_graph(axs, title, lposition, bposition):
	    ax = fig.add_subplot(axs)
	    ax.plot(theta, y, 'b-')
	    ax.set_title(title)
	    ax.spines['left'].set(position=lposition)
	    ax.spines['right'].set(visible=False)
	    ax.spines['bottom'].set(position=bposition)
	    ax.spines['top'].set(visible=False)
	    ax.xaxis.set(ticks_position='bottom')
	    ax.yaxis.set(ticks_position='left')


	theta = np.linspace(0, 2*np.pi, 128)
	y = np.sin(theta)

	fig = plt.figure(figsize=(10,6)) 
	
	ax1 = fig.add_subplot(2, 3, 1)
	ax1.plot(theta, np.sin(theta), 'b-')
	ax1.set_title('既定値')

	plot_graph(232, '軸交点を中央にする', 'center', 'center') 
	plot_graph(233, '軸交点を(0,0)にする', 'zero', 'zero')
	plot_graph(234, '軸交点をx軸25%, y軸25%にする', ('axes', 0.25), ('axes', 0.25))
	plot_graph(235, '軸交点を(1, -1)にする', ('data', 1.0), ('data', -1.0))
	plot_graph(236, '軸を離す', ('outward', 10), ('outward', 10))

	plt.tight_layout()

	plt.show();

if __name__=='__main__':
	basic()
```

# 軸目盛のカスタマイズ_axes-set_major_xxxx

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	fig = plt.figure(figsize=(10, 8))
	ax1 = fig.add_subplot(221)
	ax2 = fig.add_subplot(222)
	ax3 = fig.add_subplot(223)
	ax4 = fig.add_subplot(224)

	# 目盛なし
	ax1.plot([0, 5])
	ax1.xaxis.set(major_locator=mpl.ticker.NullLocator())
	ax1.yaxis.set(major_locator=mpl.ticker.NullLocator())	

	# パーセント表示
	ax2.plot([0, 1])
	ax2.xaxis.set(major_formatter=mpl.ticker.PercentFormatter(1))
	ax2.yaxis.set(major_formatter=mpl.ticker.PercentFormatter(10, 2, 'パーセント'))

	# フォーマット文字列指定
	ax3.plot([0, 10],[-1000000, 1000000])
	ax3.xaxis.set(major_formatter=mpl.ticker.StrMethodFormatter('{pos}月'))
	ax3.yaxis.set(major_formatter=mpl.ticker.StrMethodFormatter('{x:+,.0f}'))

	# 目盛を自由に設定
	tick_l = [0, 354, 457, 502, 525]
	label_l = ['東京', '名古屋','京都', '大阪', '神戸']
	ax4.set(
	    title='東京-神戸 間',
	    xlabel='距離(km)',
	    xticks=tick_l,
	    yticklabels=[],
	)
	ax4.set_xticklabels(label_l, rotation=45, ha='right')
	
	plt.suptitle('軸目盛のカスタマイズ')

	plt.tight_layout(pad=3, w_pad=1.0, h_pad=1.0)
	plt.show()

if __name__=='__main__':
	basic()
```

# 軸の共有_axes-twinx_axes-twiny

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	fig = plt.figure(figsize=(10,6))
	ax1 = fig.add_subplot(121)
	ax2 = ax1.twinx()        # share x axis

	ax3 = fig.add_subplot(122)
	ax4 = ax3.twiny()        # share y axis

	x = [0, 1, 2, 3, 4, 5]
	y = [19, 12, 6, 4, 3, 2]
	y1 = [41, 67, 80, 89, 96, 100]

	b = ax1.bar(x, y)
	l, = ax2.plot(x, y1, color='g', lw=3)

	ax1.set(
	    xlabel='欠陥理由コード', 
	    ylabel='欠陥数'
	)

	ax2.set(
	    ylabel='累計(%)',
	    ylim=(0, 110)
	)
	ax1.legend([b, l], ['欠陥数', '累計(%)'], loc=5)

	b1 = ax3.barh(x, y)
	l1, = ax4.plot(y1, x, color='g', lw=3)

	ax3.set(
	    xlabel='欠陥数', 
	    ylabel='欠陥理由コード',
	    ylim=(-1, 6)
	)
	ax4.set(
	    xlabel='累計(%)',
	    xlim=(0, 110)
	)
	ax3.legend([b1, l1], ['欠陥数', '累計(%)'], loc=9)

	plt.suptitle('製品の欠陥について(2020年5月)', fontsize=20)
	plt.tight_layout(w_pad=5, pad=3)

	plt.show()

if __name__=='__main__':
	basic()
```

# 色の指定_color_colorname

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	fig = plt.figure()
	ax = fig.add_subplot()

	x = np.linspace(0, 10, 20)

	color_list = [
	    'xkcd:sky blue',   # xkcd形式
	    'green',           # CSS4
	    '#1F1F1F1F',       # RGBA
	    'b',               # CSS4の略語
	    '#1C0B2D',         # RGB
	    'tab:pink',        # Tableau(タブロー)
	    'C1',              # カラーサイクル（色相環）
	]

	for i, color in enumerate(color_list):
	    y = x - (-5*i + 15)
	    ax.plot(x, y, color)
	    
	ax.set(title='カラー 例')

	plt.show()

if __name__=='__main__':
	basic()
```
