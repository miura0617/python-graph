# 1つのプロットエリアに複数のグラフ

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	x = np.linspace(-2, 5, 71) # 区間[-2, 5]を70等分する71点
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	ax.plot(x, np.sin(x), label='sin(x)', color='red')
	ax.plot(x, np.log(x), label='log(x)', color='orange')
	ax.plot(x, np.sqrt(x), label='sqrt(x)', color='blue')
	ax.plot(x, 1/x, label='1/x', color='green')
	ax.plot(x, x*x, label='x^2', color='purple')
	ax.axvline(x=0, linewidth=0.5, color='grey')
	ax.axhline(y=0, linewidth=0.5, color='grey')
	
	# プロットエリアの装飾
	ax.set(
		title='色々な線',
		xlabel='x label',
		ylabel='y lbbel',
		xlim=(-2, 5),
		ylim=(-2.5, 2.5)
	)
	ax.legend(loc='best')

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	japanese = [50, 60, 80, 70, 65]
	math = [90, 80, 85, 70, 90]
	idx = np.linspace(0, 4, 5)
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	width = 0.35
	ax.bar(idx, japanese, width, label='国語', color='lightblue')
	ax.bar(idx+width, math, width, label='数学', color='orange')
	
	# プロットエリアの装飾
	ax.set(
		title='国語と数学の得点',
		xlabel='試験回数',
		ylabel='得点(点)',
		xticks=idx + width / 2,
		xticklabels=[str(i)+"回目" for i in range(1, 6)]
	)
	ax.legend(loc='best')

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# 図(Figure)を複数のプロットエリア(axes)に分割

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	
	ax1 = fig.add_subplot(221)
	ax2 = fig.add_subplot(222)
	ax3 = fig.add_subplot(212)

	# グラフの作成
	ax1.plot([1, 2])
	ax2.plot([2, 1])
	ax3.plot([2, 4])
	
	# プロットエリアの装飾
	ax1.set(title='分割1')
	ax2.set(title='分割2')
	ax3.set(title='分割3')
	
	plt.suptitle("分割の例")
	plt.tight_layout(pad=3, w_pad=1.0, h_pad=1.0)

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# 1つのプロットエリアに複数のグラフ_gridspecを使う

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	
	gs = fig.add_gridspec(3, 3)
	
	ax1 = fig.add_subplot(gs[0,0])
	ax2 = fig.add_subplot(gs[0,1:])
	ax3 = fig.add_subplot(gs[1:, 0:2])
	ax4 = fig.add_subplot(gs[1:, 2])

	# グラフの作成
	ax1.plot([0, 1])
	ax2.plot([1, 0])
	ax3.plot([0, 2])
	ax4.plot([3, 0])
	
	# プロットエリアの装飾
	ax1.set(title='分割1')
	ax2.set(title='分割2')
	ax3.set(title='分割3')
	ax4.set(title='分割4')
	
	plt.suptitle("分割の例")
	plt.tight_layout(pad=3, w_pad=1.0, h_pad=1.0)

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# 練習_図(Figure)を複数のプロットエリア(axes)に分割

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def hist(ax):

    grade = pd.read_csv('grade.csv')
    nbins = 10
    ax.hist(grade['得点'], bins=nbins)
    ax.set(
        title='数学の得点分布',
        xlabel='得点',
        ylabel='人数',
    )
    ax.axvline(x=grade['得点'].mean(),
        linewidth=2, color='r')
    
def scatter(ax):

    medical_check = pd.read_csv('medical_check.csv', delimiter=',')

    x = medical_check['身長']
    y = medical_check['体重']

    ax.scatter(x, y)
    ax.set(
        title='身長と体重の分布',
        xlabel='身長',
        ylabel='体重',
    )
	
def stackplot(ax):

    x = np.array([i for i in range(1, 7)], dtype=np.int32)

    Apr = [5, 7, 6, 8, 7, 9]
    May = [0, 4, 3, 7, 8, 9]
    June = [6, 7, 4, 5, 6, 8]

    labels = ["4月", "5月", "6月"]

    ax.stackplot(x, Apr, May, June, labels=labels)
    ax.legend(loc=2)

    ax.set(
        title='プロダクト品質 2020/1Q',    
        xlabel='原因コード',
        ylabel='発生件数',
    )
	
def pie(ax):

    labels = ['ＳＦ', 'ドラマ', 'スリラー', 'コメディー', 'アクション', 'ロマンス']
    sizes = [5, 15, 10, 20, 40, 10]

    explode = (0, 0, 0, 0, 0.1, 0)

    ax.pie(sizes, labels=labels, explode=explode, counterclock=False, 
        autopct='%1.1f%%', shadow=True, startangle=90)

def basic():

	# データの用意
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure(figsize=(8, 8))
	
	ax1 = fig.add_subplot(221)
	ax2 = fig.add_subplot(222)
	ax3 = fig.add_subplot(223)
	ax4 = fig.add_subplot(224)
	
	
	# グラフの作成
	## ヒストグラムの表示
	hist(ax1)
	
	## 散布図
	scatter(ax2)
	
	## 積み上げグラフ
	stackplot(ax3)
	
	## 円グラフ
	pie(ax4)
	
	# プロットエリアの装飾
	
	plt.suptitle("分割表示の例")
	plt.tight_layout(pad=3, w_pad=1.0, h_pad=1.0)

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```
