# Figure(図)とAxes(プロットエリア)の関係

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rc('font', family="MS Gothic")

def intro():

	x = np.linspace(0, 5, 51)
	y = x * x
	
	fig = plt.figure()
	ax = fig.add_subplot()
	ax.plot(x, y)
	
	ax.set_title('2次曲線')
	ax.set_xlabel('x座標')
	ax.set_ylabel('y座標')
	plt.show()

if __name__=='__main__':
	intro()
```

# 棒グラフ(縦)bar

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def intro():

	month = [ str(i)+"月" for i in range(1, 13)]
	sold  = [ 55, 45, 50, 60, 75, 90, 110, 105, 100, 95, 80, 70]
	
	fig = plt.figure()
	ax = fig.add_subplot()
	bar_rects = ax.bar(month, sold)
	
	for rect in bar_rects:
		x = rect.get_x()
		y = rect.get_height()
		width = rect.get_width()
		ax.text(x+width/2, 1.002*y, int(y), ha='center', va='bottom')
	
	ax.set(
		title='アイスコーヒー売上',
		ylabel='売上高(千円)'
	)
	plt.show()

if __name__=='__main__':
	intro()
```

# 棒グラフ(横)barh

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def intro():

	month = [ str(i)+"月" for i in range(1, 13)]
	sold  = [ 55, 45, 50, 60, 75, 90, 110, 105, 100, 95, 80, 70]
	
	fig = plt.figure()
	ax = fig.add_subplot()
	bar_rects = ax.barh(month, sold)
	
	for rect in bar_rects:
		x = rect.get_width()
		y = rect.get_y()
		height = rect.get_height()
		ax.text(1.002*x, y+height/2, int(x), ha='left', va='center')
	
	ax.set(
		title='アイスコーヒー売上',
		xlabel='売上高(千円)'
	)
	plt.show()

if __name__=='__main__':
	intro()
```

# 散布図scatter

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	medical_check = pd.read_csv('medical_check.csv', delimiter=',')
	
	x = medical_check['身長']
	y = medical_check['体重']

	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	ax.scatter(x, y)
	ax.grid()

	# プロットエリアの装飾
	ax.set(
		title='身長と体重',
		xlabel='身長',
		ylabel='体重'
	)
	
	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# 積み上げグラフ_stackplot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	x = [i for i in range(1, 7)]
	
	Apr = [5, 7, 6, 8, 7, 9]
	May = [0, 4, 3, 7, 8, 9]
	June = [6, 7, 4, 5, 6, 8]
	
	labels = ["4月", "5月", "6月"]
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	ax.stackplot(x, Apr, May, June, labels=labels)

	# プロットエリアの装飾
	ax.set(
		title='プロダクト品質 2020/1Q',
		xlabel='原因コード',
		ylabel='発生件数'
	)
	ax.legend(loc=2)
	
	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# 円グラフ_pie

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	sizes = [10, 30, 20, 40, 80, 20]
	labels = ['ＳＦ', 'ドラマ', 'スリラー', 'コメディー', 'アクション', 'ロマンス']
	explode = [0, 0, 0, 0, 0.1, 0]
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	ax.pie(
		sizes,
		labels=labels,
		explode=explode,
		counterclock=False,
		autopct='%1.1f%%',
		shadow=True,
		startangle=90
	)
	
	# プロットエリアの装飾
	#ax.set(
	#	title='好きな映画',
	#)
	ax.set_title('好きな映画', fontsize=16)
	
	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```

# ヒストグラム_hist

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

mpl.rc('font', family="MS Gothic")

def basic():

	# データの用意
	grade = pd.read_csv('grade.csv')
	
	# 図(fig)、プロットエリア(axes)の生成
	fig = plt.figure()
	ax = fig.add_subplot()
	
	# グラフの作成
	ax.hist(grade['得点'], bins=10)
	ax.axvline(x=grade['得点'].mean(), linewidth=2, color='r')
	
	# プロットエリアの装飾
	ax.set(
		title='数学の得点分布',
		xlabel='得点',
		ylabel='人数'
	)

	# 図の描画
	plt.show()

if __name__=='__main__':
	basic()
```
